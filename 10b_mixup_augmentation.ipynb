## Mixup

### What is mixup?

As the name kind of suggests, the authors of the  [mixup article](https://arxiv.org/abs/1710.09412) propose to train the model on a mix of the pictures of the training set. Let's say we're on CIFAR10 for instance, then instead of feeding the model raw images, we take two(which could be in the same class or not) and do a linear combination of them: in terms of tensor it's
```python
new_image = t* image1 + (1-t) * image2
```
where t is a float between 0 and 1. Then the target we assign to that image is the same combination of the original targets:
```python
new_target = t* target1 + (1-t) * target2
```
assuming the targets are one-hot encoded(which isn't the case with pytorch usually).  
  
Basically mixup aids the model in forming better decision boundaries between classes. It is also really effective in making the model immune to adversarial inputs. You can train for a longer time. Really efficient if you need more data without actually collecting any. Overall, it helps the model generalize better.

In the original article, the author suggested three things:  
- Create two seperate dataloaders and draw a batch from each at every iteration to mi them up.
- Draw a t value following a beta distribution with a parameter ( 0.4 is suggested in their article).
- Mix up the two batches with the same value t.
- Use one hot encoded targets.

With a low ``, we pick values close to 0. and 1. with a higher probability, and the values in the middle all have the same kind of probability. With a greater ``, 0. and 1. get a lower probability.

While the approach above works very well, it's not the fastest way we can do this. The main point that slows down this process is wanting two different batces at every iteration( which means loding twice the amount of images and applying them the other data augmentation functions). To avoid this slowdown, we can be a little smarter and mixup a batch with a shuffled version of itself (this way images mixed up are still different). This was a trick suggested by the mixup paper.
  
Then pytorch was very careful to avoid one-hot encoding when it could, so it seems a bit of a drag to undo this. Fortunately for us, if the loss is a classic crross entropy, we have
```python
loss(output, new_target) = t* loss(output, target1) + (1-t)* loss(output, target2
``` 
so we won''t one hot encode anything and just compute those two losses then do the linear combination. 
  
Using the same parameter t for the whole batch also seemed a bit inefficient. In our experiments, we noticed that the model can train faster if we draw a different t for every imae in the batch(both options get to the same result in terms of accuracy, it's just tat one arrives there more slowly). The last trick we have to apply with this is that there can be some duplicates with this strategy: let's say or shuffle say to mix image0 and image1 then image1 with image0, and that we draw t=0.1 for the first and t=0.9 for the second. Then
```python 
image0 * 0.1 + shuffle0 * (1-0.1) = image0 * 0.1 + image1 * 0.9
image1 * 0.9 + shuffle1 * (1-0.9) = image1 * 0.9 + image0 * 0.1
```
will be the same. OF course, we'd have to be a bit unlucky but in practice, we saw there was drop in accuracy by using this without removing those near duplicates. To avoid them, the trick is to replace the vector of parameters we drew by 
```python
t = max(t, 1-t)
```
The beta distribution with the two parameters equal is symmetric in any case, and this way we ensure that the biggest coefficient is always near the first image(the non-shuffled batch).   
   
In `Mixup` we have handle loss functions that have an attribute `reduction` (like `nn.CrossEntropy()`). To deal with the `reduction=None` with various types of loss function without modifying the actual loss function outside of the scope we need to perform those operations with no reduction, we create a context manager:

````python
#export
class NoneReduce():
    def __init__(self, loss_func):
        self.loss_func, self.old_red = loss_func, None
        
    def __enter__(self):
        if hasattr(self.loss_func, 'reduction'):
            self.old_red = getattr(self.loss_func, 'reduction')
            setattr(self.loss_func, 'reduction', 'none')
            return self.loss_func
        else: return partial(self.loss_func, reduction='none')
        
    def __exit__(self, type, value, traceback):
        if self.old_red is not None: setattr(self.loss_func, 'reduction', self.old_red)
````

Then we can use it in Mixup:

````python
#export
from torch.distributions.beta import Beta

def unsqueeze(input, dims):
    for dim in listify(dims): input = torch.unsqueeze(input, dim)
    return input 
    
def reduce_loss(loss, reduction='mean'):
    return loss.mean() if reduction == 'mean' else loss.sum()
````

````python
#export
class MixUp(Callback):
    _order = 90 # Runs after normalization and cuda
    def __init__(self, :float=0.4): self.distrib = Beta(tensor([]), tensor([]))
        
    def begin_fit(self): self.old_loss_func, self.run.loss_func = self.run.loss_func, self.loss_func
    
    def begin_batch(self):
        if not self.in_train: return # Only mixup things during training
         = self.distrib.sample((self.yb.size(0),)).squeeze().to(self.xb.device)
         = torch.stack([, 1-], 1)
        self. = unsqueeze(.max(1)[0], (1,2,3))
        shuffle = torch.randperm(self.yb.size(0)).to(self.xb.device)
        xb1, self.yb1 = self.xb[shuffle], self.yb[shuffle]
        self.run.xb = lin_comb(self.xb, xb1, self.)
        
    def after_fit(self): self.run.loss_func = self.old_loss_func
        
#     We have to turn off this 'reduceloss' while we're doing mixup cause we actually 
#     need to calculate the loss on every image for both halves of our mixup. THis is why we use a context manager 
#     here which removes the reduction and puts it back when finished.
    def loss_func(self, pred, yb):
        if not self.in_train: return self.old_loss_func(pred, yb)
        with NoneReduce(self.old_loss_func) as loss_func:
            loss1 = loss_func(pred, yb)
            loss2 = loss_func(pred, self.yb1)
        loss = lin_comb(loss1, loss2, self.)
        return reduce_loss(loss, getattr(self.old_loss_func, 'reduction', 'mean'))
````

*Question*: How does softmax interact with all of this? Should we jump straight from mixup to inference?   
  
To really use mixup well, or when your data isn't all correctly labeled you don't wanna be asking your model to predict one class at with 100% confidence as you have label noise(incorrect labels, mixed up images etc). So instead we say to not use one hot encoding for the labels but use 0.9( or a little bit less than 1) hot encoding. The correct answer for the model would be to say I'm 90% sure is this label and all the probabilaties have to add to 1 so then for all of the negatives you simply put 0.1/(n-1)to all of them. This is called Label smoothing and it is a really great way to treat noisy labels.

## Label smoothing

Another regularization technique that's often used is label smoothing. It's designed to make the model a little bit less certain of it's decision by changing a little bit its target: instead of wanting to predict 1 for the correct class and 0 for all the others, we ask it to predict `1-` for the correct class and `` for all the others, with `` a (small) positive number and N the number of classes. This can be written as:

$$loss = (1-) ce(i) +  \sum ce(j) / N$$

where `ce(x)` is cross-entropy of `x` (i.e. $-\log(p_{x})$), and `i` is the correct class. This can be coded in a loss function:

Jeremy has also observed that even after people random permuting the labels and training the model with label smoothing they end up getting really good results. So we shouldn't wait to build the model until We're pretty sure that all the labels are correct. If the result turns out to be okay then you can skip the clean up work that was required.

````python
#export
class LabelSmoothingCrossEntropy(nn.Module):
    def __init__(self, :float=0.1, reduction='mean'):
        super().__init__()
        self., self.reduction = , reduction
        
    def forward(self, output, target):
        c = output.size()[-1]
        log_preds = F.log_softmax(output, dim=-1)
        loss = reduce_loss(-log_preds.sum(dim=-1), self.reduction)
        nll = F.nll_loss(log_preds, target, reduction=self.reduction)
        return lin_comb(loss/c, nll, self.)
````

Note: We implement the various reduction attributes so that it plays nicely with MixUp after.
